```python
from PIL import Image
import pandas as pd
import numpy as np
from keras.utils import np_utils
print("Imported")
n_classes = 5;
```

```python
csv = pd.read_csv("C:\\Users\\06man\\Documents\\ml_data\\train.csv")
print(csv.head())
image_count = 1100
start_count = 1000
x_data = {}
y_data = {}
base_path = "C:\\Users\\06man\\Documents\\ml_data\\"
```

```python
for index, row in csv.iterrows():
    # print(f"Index: {index} ImageId: {row['image_id']}, label: {row['label']}")
    if index >= start_count and index < image_count:
        image = Image.open(f"{base_path}train_images\\{row['image_id']}")
        image = image.resize((224,224))
        img_arr = np.array(image)
        img_arr = img_arr.astype('float32')
        img_arr /= 255
        x_data[index] = img_arr
        y_data[index] = row['label']
        print(index)
        # print(x_data)
        # print(y_data)
    elif index < image_count:
        print(f"skipping:{index}")
    else:
        print("BREAKING")
        break
    
```

```python
x_data = np.array([list(item) for item in x_data.values()])
y_data = np.array([item for item in y_data.values()])
print(x_data.shape)
print(y_data.shape)
print("Shape before one-hot encoding: ", y_data.shape)
y_data = np_utils.to_categorical(y_data, n_classes)
print("Shape after one-hot encoding: ", y_data.shape)
```

```python
np.save(f"{base_path}x_test",x_data)
np.save(f"{base_path}y_test",y_data)
print("SAVED")
```
